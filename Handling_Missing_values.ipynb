# **Visualizing Missing Data with Seaborn Heatmap and Displot**

- Heatmap takes a rectangular data grid as input and then assigns a color intensity to each data cellbased on the data value of the cell.
-This is a great way to get visual clues about the data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
sns.set_theme()
```

```python
df=pd.read_csv('data2.csv')
```

```python
plt.figure(figsize=(28,12))
plt.xticks( fontsize=20)
plt.yticks( fontsize=20)

sns.heatmap(df.corr(), cmap='Purples', annot=True, annot_kws={'size':12})
plt.title('corrolation among features', fontsize=20);
```
